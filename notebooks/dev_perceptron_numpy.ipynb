```python
import sys
sys.path.append("../src")

from data_loader import load
from preprocessor import preprocess
from perceptron_numpy import init_weights, train, predict
```

```python
# load and preprocess data
df = load(source="synthetic")
X_train, X_val, y_train, y_val = preprocess(df)
```

```python
# initialize weights
w, b = init_weights(X_train.shape[1])
print("Initial weights:", w)
print("Initial bias   :", b)
```

```python
# train
print("\nTraining...\n")
w, b, losses = train(X_train, y_train, lr=0.01, epochs=50)
```

```python
# predict on val set
y_pred = predict(X_val, w, b)
```

```python
# accuracy check
accuracy = (y_pred == y_val).mean()
print(f"\nVal accuracy: {accuracy:.4f}")
```
