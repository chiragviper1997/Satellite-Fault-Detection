```python
import numpy as np
import sys
sys.path.append("../src")

from data_loader import load
from preprocessor import preprocess
from perceptron_numpy import train, predict
from utils import set_seed, save_weights, load_weights

# set seed
set_seed(42)

# load and preprocess
df = load(source = "synthetic")
X_train, X_val, y_train, y_val = preprocess(df)

# train
w, b, losses = train(X_train, y_train, lr = 0.01, epochs = 50)

# save weights
save_weights(w, b)

# load weights back
w_loaded, b_loaded = load_weights()

# verify loaded weights match original
print("\nWeights match:", np.allclose(w, w_loaded))
print("Bias matches  :", np.isclose(b, b_loaded))
```
