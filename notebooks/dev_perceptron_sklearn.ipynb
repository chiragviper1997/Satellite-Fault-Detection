```python
import sys
sys.path.append("../src")

from data_loader import load
from preprocessor import preprocess
import perceptron_sklearn as sk_slp

# load and preprocess
df = load(source = "synthetic")
X_train, X_val, y_train, y_val = preprocess(df)

# train
print("\nTraining...\n")
model, _, losses = sk_slp.train(X_train, y_train, lr = 0.01, epochs = 50)

# predict on val set
y_pred = sk_slp.predict(X_val, model)

# accuracy
accuracy = (y_pred == y_val).mean()
print(f"\nVal accuracy: {accuracy:.4f}")
```
